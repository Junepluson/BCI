```python
import os
import scipy.io
import matplotlib.pyplot as plt

# DEAP    
folder_path = '../DEAP'

#  .mat   
mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

# Valence    
valence_data_per_video = {i: [] for i in range(40)}  #    (40 )

#  .mat   Valence  
for mat_file in mat_files:
    file_path = os.path.join(folder_path, mat_file)
    mat_data = scipy.io.loadmat(file_path)
    
    #   Valence   (labels   )
    valence_labels = mat_data['labels'][:, 0]  # 40   Valence 
    for video_index, valence in enumerate(valence_labels):
        valence_data_per_video[video_index].append(valence)

#   Valence   
plt.figure(figsize=(20, 10))
for video_index, valences in valence_data_per_video.items():
    plt.subplot(5, 8, video_index + 1)  # 5 8  
    plt.hist(valences, bins=10, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f"Video {video_index + 1}")
    plt.xlabel("Valence")
    plt.ylabel("Frequency")
    plt.tight_layout()

plt.suptitle("Valence Distribution Across Videos for All Participants", y=1.02, fontsize=16)
plt.show()
```

```python
import os
import scipy.io
import matplotlib.pyplot as plt

# DEAP    
folder_path = '../DEAP'

#  .mat   
mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

#  (40)  32 Valence Arousal    
valence_arousal_data_per_video = {i: {'valence': [], 'arousal': []} for i in range(40)}

#  .mat   Valence Arousal  
for mat_file in mat_files:
    file_path = os.path.join(folder_path, mat_file)
    mat_data = scipy.io.loadmat(file_path)
    
    #   Valence Arousal  
    valence_labels = mat_data['labels'][:, 0]  # Valence  (40)
    arousal_labels = mat_data['labels'][:, 1]  # Arousal  (40)
    for video_index in range(40):
        valence_arousal_data_per_video[video_index]['valence'].append(valence_labels[video_index])
        valence_arousal_data_per_video[video_index]['arousal'].append(arousal_labels[video_index])

#   Valence Arousal  
plt.figure(figsize=(20, 15))
for video_index, data in valence_arousal_data_per_video.items():
    plt.subplot(5, 8, video_index + 1)  # 5 8  
    plt.scatter(data['valence'], data['arousal'], alpha=0.7, color='blue')
    plt.title(f"Video {video_index + 1}")
    plt.xlabel("Valence")
    plt.ylabel("Arousal")
    plt.tight_layout()

plt.suptitle("Valence vs Arousal for All Videos", y=1.02, fontsize=16)
plt.show()
```

```python
import os
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# DEAP    
folder_path = '../DEAP'

#  .mat   
mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

#    Valence Arousal  
plt.figure(figsize=(20, 15))

for person_index, mat_file in enumerate(mat_files):
    file_path = os.path.join(folder_path, mat_file)
    mat_data = scipy.io.loadmat(file_path)
    
    # Valence Arousal  
    valence_labels = mat_data['labels'][:, 0]  # Valence  (40)
    arousal_labels = mat_data['labels'][:, 1]  # Arousal  (40)
    
    #   ([0, 1] )
    scaler = MinMaxScaler()
    valence_labels_normalized = scaler.fit_transform(valence_labels.reshape(-1, 1)).flatten()
    arousal_labels_normalized = scaler.fit_transform(arousal_labels.reshape(-1, 1)).flatten()
    
    #    
    plt.subplot(8, 4, person_index + 1)  # 8 4  
    plt.scatter(valence_labels_normalized, arousal_labels_normalized, alpha=0.7, color='blue')
    plt.title(f"Person {person_index + 1}")
    plt.xlabel("Normalized Valence")
    plt.ylabel("Normalized Arousal")
    plt.tight_layout()

plt.suptitle("Normalized Valence vs Arousal for All Participants", y=1.02, fontsize=16)
plt.show()
```

```python
import os
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# DEAP    
folder_path = '../DEAP'

#  .mat   
mat_files = [f for f in os.listdir(folder_path) if f.endswith('.mat')]

# Valence Arousal   
valence_means, valence_stds = [], []
arousal_means, arousal_stds = [], []

#   Valence Arousal   ,  
for mat_file in mat_files:
    file_path = os.path.join(folder_path, mat_file)
    mat_data = scipy.io.loadmat(file_path)
    
    # Valence Arousal  
    valence_labels = mat_data['labels'][:, 0]  # Valence  (40)
    arousal_labels = mat_data['labels'][:, 1]  # Arousal  (40)
    
    #  
    scaler = MinMaxScaler()
    valence_labels_normalized = scaler.fit_transform(valence_labels.reshape(-1, 1)).flatten()
    arousal_labels_normalized = scaler.fit_transform(arousal_labels.reshape(-1, 1)).flatten()
    
    #   
    valence_means.append(np.mean(valence_labels_normalized))
    valence_stds.append(np.std(valence_labels_normalized))
    arousal_means.append(np.mean(arousal_labels_normalized))
    arousal_stds.append(np.std(arousal_labels_normalized))

#   (Valence vs Arousal )
plt.figure(figsize=(10, 8))
plt.errorbar(valence_means, arousal_means, xerr=valence_stds, yerr=arousal_stds, fmt='o', capsize=5, color='blue', ecolor='gray', elinewidth=1)
plt.title("Bias Analysis: Normalized Valence vs Arousal", fontsize=16)
plt.xlabel("Mean Valence (with Std Dev)", fontsize=12)
plt.ylabel("Mean Arousal (with Std Dev)", fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
